```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
results = pickle.load(open('spatial_cat_1_spatialDMS_WTLoss.pkl','rb'))
plt.imshow(results['weights']['w_in'].T)
plt.colorbar()
#plt.savefig('w_in.png',bbox_inches='tight')
plt.xlabel('Hidden neurons')
plt.ylabel('Input neurons')
plt.title('Trained on spatial working memory and spatial cat')
plt.show()
```

```python
wt=results['weights']['w_in']
wt_summed=np.sum((wt),axis=0)
wt_sumrshp=np.reshape(wt_summed[:70],(7,10))
wt_space=np.flipud(wt_sumrshp.T)
plt.imshow(wt_space)
plt.colorbar()
#plt.savefig('w_in.png',bbox_inches='tight')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Trained on spatial working memory and spatial cat only')
plt.show()
```
